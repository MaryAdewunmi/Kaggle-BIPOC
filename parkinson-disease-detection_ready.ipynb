### Introduction

Parkinson disease (PD) is one of the most common neurologic diseases, affecting about 1% of people over the age of 60 and resulting in progressive impairment that can be delayed but not stopped with treatment. The absence of pigmented dopaminergic neurons in the substantia nigra pars compacta, as well as the presence of Lewy bodies and Lewy neurites, are the two most common neuropathologic findings in Parkinson's disease(https://emedicine.medscape.com/article/1831191-overview)

It is a degenerative neurological condition characterized by low levels of dopamine in the brain. It shows up as a deterioration in movement, including tremors and stiffness. Dysarthria (difficulty articulating sounds), hypophonia (lower volume), and monotone speech are all common side effects (reduced pitch range). Additionally, cognitive impairments and mood fluctuations are possible, as well as an increased risk of dementia.Traditional Parkinson's disease diagnosis entails a clinician taking a patient's neurological history and examining motor skills in a variety of conditions. Since there is no definitive laboratory test for PD, it can be difficult to diagnose, particularly in the early stages when motor symptoms are mild. The patient must return to the clinic on a regular basis to monitor the disease's progression. It would be beneficial to have an efficient screening process, particularly one that does not require a clinic visit. Since PD patients have distinct vocal characteristics, voice recordings are a valuable and non-invasive diagnostic tool. If machine learning algorithms could reliably diagnose Parkinson's disease using a voice recording dataset, this would be a useful screening phase before seeing a clinician('Exploiting Nonlinear Recurrence and Fractal Scaling Properties for Voice Disorder Detection',
Little MA, McSharry PE, Roberts SJ, Costello DAE, Moroz IM.
BioMedical Engineering OnLine 2007, 6:23 (26 June 2007)).

Attribute Information:
Matrix column entries (attributes):
name - ASCII subject name and recording number
MDVP:Fo(Hz) - Average vocal fundamental frequency
MDVP:Fhi(Hz) - Maximum vocal fundamental frequency
MDVP:Flo(Hz) - Minimum vocal fundamental frequency
MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Several

MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude

Setting gamma value to auto is the best because a lower value of gamma will loosely fit the training dataset causing underfitting, whereas a higher value of gamma will exactly fit the training dataset, which causes over-fitting. In other words, you can say a low value of gamma considers only nearby points in calculating the separation line, while the a value of gamma considers all the data points in the calculation of the separation line.

Setting kernel= linear because we have just 2 results which are 0 & 1- Either the patient has parkinson disease(1) or not(0)

* False Postive(FP) = 3 - We predicted yes, but they don't actually have the disease. (Also known as a "Type I error.")
* True Positive(TP) = 29 -These are cases in which we predicted yes (they have the disease), and they do have the disease.
